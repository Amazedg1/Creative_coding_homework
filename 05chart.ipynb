```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
path = 'Headphone.csv'
```

```python
headphone = pd.read_csv(path)
```

```python
frequency_counts = headphone['How frequently do you use a headset?'].value_counts()
frequency_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'],fontsize =30)
plt.title('Frequency of Headphone Usage')
plt.ylabel('')
plt.show()
#Looks like headphones are full-time items now
```

```python
wired_wireless_counts = headphone['Do you prefer wired or wireless headsets?'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Most of CTA group are not old school guys')
plt.ylabel('number of people')
plt.show()
#I am really curious about the
```

```python
plt.hist(headphone['ow much are you willing to spend on a high-quality headset?'], bins=6)
plt.ylabel('number of people')
plt.title('how much are you willing to spend on a high-quality headset?')
plt.show()
#
```

```python
plt.hist(headphone['What brand of headsets do you currently own or prefer?'], bins=10,)
plt.ylabel('Number of people')
plt.title('What brand CTA use')
plt.show()
```

```python
bang_olufsen_cta = headphone[headphone['What brand of headsets do you currently own or prefer?'] == 'Bang_&_Olufsen']
```

```python
wide=10
a=0.5
c='Which factor is most important to you when choosing a headset?'

plt.hist(apple_cta[[c]], bins=wide, label='Apple', alpha=a)
plt.hist(sony_cta[[c]], bins=wide, label='Sony', alpha=a)
plt.hist(Sennheiser_cta[[c]], bins=wide, label='Sennheiser', alpha=a)
plt.hist(bose_cta[[c]], bins=wide, label='Bose', alpha=a)
plt.hist(bang_olufsen_cta[[c]], bins=wide, label='Bang_&_Olufsen', alpha=a)


plt.xlabel('Factors Importance', fontsize= 8)
plt.ylabel('Frequency')
plt.title('Histogram of Factors Importance for Different Headset Brands')
plt.legend()


plt.show()


```

```python
brands = ['Apple', 'Sony', 'Sennheiser', 'Bose', 'Bang & Olufsen']

fig, axes = plt.subplots(1, 5, figsize=(15, 3), subplot_kw=dict(aspect="equal"))
#two variables, creat subplots, 1 row,5 column, the hole chart size wide 15 hgih 3, make the chart ratio equal

for i, brand in enumerate(brands):
        #the index of current barnd, enumerate return the both the value and index in the list, for loop get the index
    brand_data = headphone[headphone['What brand of headsets do you currently own or prefer?'] == brand]
        #get the brand choices data from column brand
    factors_importance = brand_data['How important is the design and aesthetics of a headset to you?'].value_counts()
        #get the value_counts from column design
    
    axes[i].pie(factors_importance, labels=factors_importance.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
    # creat pie chart use the value in factor importance, creat the labels as the index of , use %
    axes[i].set_title(brand)

fig.suptitle('Design and aesthetics Importance for Different Headset Brands users in CTA', fontsize=16)
#title
plt.show()
```

```python
brands = ['Apple', 'Sony', 'Sennheiser', 'Bose', 'Bang & Olufsen']

fig, axes = plt.subplots(1, 5, figsize=(15, 3), subplot_kw=dict(aspect="equal"))

for i, brand in enumerate(brands):
    brand_data = headphone[headphone['What brand of headsets do you currently own or prefer?'] == brand]
    factors_importance = brand_data['Which additional features would you like in a headset? (Select up to two)'].value_counts()
    
    axes[i].pie(factors_importance, labels=factors_importance.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
    axes[i].set_title(brand)

fig.suptitle('Additional features Importance for Different Headset Brands users in CTA', fontsize=16)
plt.show()
```

```python
brands = ['Apple', 'Sony', 'Sennheiser', 'Bose', 'Bang & Olufsen']

fig, axes = plt.subplots(1, 5, figsize=(15, 3), subplot_kw=dict(aspect="equal"))

for i, brand in enumerate(brands):
    brand_data = headphone[headphone['What brand of headsets do you currently own or prefer?'] == brand]
    factors_importance = brand_data['What feature do you prioritize for communication headsets (e.g., for calls or online meetings)?'].value_counts()
    
    axes[i].pie(factors_importance, labels=factors_importance.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
    axes[i].set_title(brand)

fig.suptitle('Factors Importance for Different Headset Brands users in CTA', fontsize=16)
plt.show()
```
